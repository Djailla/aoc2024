```python
with open("input.txt") as f:
    data = f.read().splitlines()
```

```python
from itertools import product
from functools import reduce

def get_combinations(size):
    # Génération des combinaisons
    return list(product(["+", "x"], repeat=size))

def merge_list(numbers, operators):
    out = [numbers[0]]
    for idx in range(len(operators)):
        out.append(operators[idx])
        out.append(numbers[idx + 1])

    return out
```

```python
data_dict = []

for l in data:
    res, numbers = l.split(": ")
    data_dict.append((int(res), [int(n) for n in numbers.split(" ")]))
```

```python
from itertools import groupby

def split_list(line, sep):
    res = []
    
    for key, group in groupby(line, lambda x: x != sep):
        if key:  
            res.append(list(group))

    return res
```

```python
tot = 0

import operator
def prod(l):
    return reduce(operator.mul, [e[0] for e in l], 1)

for target, numbers in data_dict:
    size = len(numbers) - 1
    combinations = get_combinations(size)
    for c in combinations:
        m = merge_list(numbers, c)
        res = sum([prod(split_list(e, "x")) for e in split_list(m, "+")])
        if res == target:
            print(m)
```

```python
import operator

def get_operator_combinations(size):
    # Génération des combinaisons
    return list(product([operator.add, operator.mul], repeat=size))

def compute_result(numbers, operators):
    out = numbers[0]
    for idx in range(len(operators)):
        out = operators[idx](out, numbers[idx + 1])
    return out

tot = 0

for target, numbers in data_dict:
    size = len(numbers) - 1
    combinations = get_operator_combinations(size)
    for c in combinations:
        m = compute_result(numbers, c)
        if m == target:
            print(target, numbers, c)
            tot += target
            break

tot
```
